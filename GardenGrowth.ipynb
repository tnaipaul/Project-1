coneflowers growth rate:0.02 in rainy season, -0.01 in dry season, if % of gradens total area that is sunny falls below 70%, growth rate is 0.1 lower than otherwise, if % falls below 50%, growth rate is 0.2 lower

hostas growth rate: -0.01 in rainy season, 0.02 in dry season

sedums growth rate: -0.01 in rainy season, 0.03 in dry season, if the sunny area % falls below 50%, growth rate is 0.05 lower than otherwise

ferns growth rate: 0.02 in rainy season, -0.01 in dry season

ornamental trees(saplings), after 4 years they will be large enough to produce shade, assume tree has 0 canopy until 4 yrs

canopy, starting at age 4, radius growth rate: 0.8 ft/yr until age 14 (when tree ceases to grow)

sun blockage use area of circle, assume no two trees' canopies overlap

rainy season January-June (181 days)

dry season July-December (184 days)

carrying capacity 1 plant per square foot (doesn't apply to trees)

```
conefls_pop = 12 #initial population
hostas_pop = 19 #initial
sedums_pop = 14  #initial
ferns_pop = 7 #initial
trees = 12 #saplings
total_time = 20*12 #growth period (months)
deltaT = 1/30 #months
N = int(total_time/deltaT)
area = 1600 #area of the garden in sq ft
cone_r = 0 #growth rate of conefls 
hosta_r = 0 #growth rate of hostas 
sedum_r = 0 #growth rate of sedums
fern_r = 0 #growth rate of ferns
radius_rate = 0.8 #ft/yr
canopy = 0 #radius of canopy
covered_area = 0 #area covered by canopy
percentage = 100 #percentage of sunny garden

capacity = 1600 #plants




import math


for i in range(1, N+1): #runs the for loop N amount of times
  t = i*deltaT #calculates the time t in months



 
 
  if t > 48 and t <= 168 and t%12 == 0: #checks if 4 years (48 months) have passed and it's less than 14 years (168 months)
    canopy += radius_rate #incresing the size of the canopy by 0.8
    covered_area = trees*(math.pi*(canopy**2)) #calculating the size of the area covered in shade
  percentage = 100 - ((covered_area/area)*100) #calculates the percentage of the garden that is still sunny





  if t%12 <= 6: #checks if it's the rainy season
    cone_r = 0.02
    hosta_r = -0.01
    sedum_r = -0.01
    fern_r = 0.02
  else: #dry season
    cone_r = -0.01
    hosta_r = 0.02
    sedum_r = 0.03
    fern_r = -0.01






  if percentage < 50: #checks if the canopy covers more than 50% of the garden (less than 50% of the garden is sunny)
    conefls_pop += ((cone_r-.2)*((capacity-conefls_pop)/capacity)*conefls_pop)*deltaT #logistic equation to calculate the population
    hostas_pop += (hosta_r*((capacity-hostas_pop)/capacity)*hostas_pop)*deltaT 
    sedums_pop += ((sedum_r-.05)*((capacity-sedums_pop)/capacity)*sedums_pop)*deltaT 
    ferns_pop += (fern_r*((capacity-ferns_pop)/capacity)*ferns_pop)*deltaT 
    total_plants = conefls_pop + hostas_pop + sedums_pop + ferns_pop #calculates the total number of plants in the garden
  elif percentage < 70: #checks if the canopy covers 30-50% of the garden (garden is 50-70% sunny)
    conefls_pop += ((cone_r-.1)*((capacity-conefls_pop)/capacity)*conefls_pop)*deltaT
    hostas_pop += (hosta_r*((capacity-hostas_pop)/capacity)*hostas_pop)*deltaT 
    sedums_pop += ((sedum_r)*((capacity-sedums_pop)/capacity)*sedums_pop)*deltaT
    ferns_pop += (fern_r*((capacity-ferns_pop)/capacity)*ferns_pop)*deltaT
    total_plants = conefls_pop + hostas_pop + sedums_pop + ferns_pop
  else: #canopy covers less than 30% of the garden (more than 70% of the garden is sunny)
    conefls_pop += ((cone_r)*((capacity-conefls_pop)/capacity)*conefls_pop)*deltaT
    hostas_pop += (hosta_r*((capacity-hostas_pop)/capacity)*hostas_pop)*deltaT 
    sedums_pop += ((sedum_r)*((capacity-sedums_pop)/capacity)*sedums_pop)*deltaT 
    ferns_pop += (fern_r*((capacity-ferns_pop)/capacity)*ferns_pop)*deltaT 
    total_plants = conefls_pop + hostas_pop + sedums_pop + ferns_pop
  print(total_plants, t)
  
```
